```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
from scipy.optimize import curve_fit
from scipy.stats import chi2
import scienceplots
plt.style.use(['science','notebook','grid'])
import openpyxl as opx

path = np.loadtxt(r"C:\Users\giuse\OneDrive\Desktop\PythonLAB\AzotoBiatomico\Hg_1_txt.txt")

x1 = np.zeros(2047) #Wavelenghts
y1 = np.zeros(2047) # Intensit√†

wv_Hg = [] #Lunghezze d'onda righe mercurio
s_Hg = []  # Errore lunghezze d'onda

print(len(path))

for i in range (2047):
    
    x1[i] = path[i][0]
    y1[i] = path[i][1]
    

y1 = y1 / 1000
print(x1)
print(y1)

plt.figure(figsize=(5,5))
plt.scatter(x1 , y1)
```

<b> RIGA 253.6 nm

```python
plt.figure(figsize=(5,5))
plt.scatter(x1[130:152], y1[130:152] , marker='.') 
x1_fit = x1[130:152]
y1_fit = y1[130:152]
```

```python
def fit_func( x , A , k  , u  , q):
    return A * np.exp(-k*(x-u)**2)  + q
guess =np.array([60 ,  0.16 ,  253.6 , 5 ])
popt , pcov = curve_fit(fit_func , x1_fit , y1_fit  ,  absolute_sigma =True  ,p0 = guess )
perr = np.sqrt(np.diag(pcov))
print(popt)
print(perr)

print("Ampiezza = {:.6f} ¬± {:.6f}".format(popt[0], perr[0]))
print("Lunghezza d'onda picco  = {:.6f} ¬± {:.6f}".format(popt[2], perr[2]))

wv_Hg.append(popt[2])
s_Hg.append(perr[2])

x_model = np.linspace( min(x1[120:170]) , max(x1[120:170]) , 1000)
y_model = fit_func (x_model , *popt ) 

plt.figure(figsize=(6,4))
plt.plot( x_model , y_model , color='k' )
plt.scatter(x1[120:170] , y1[120:170] , marker ='.' , label = 'Misure')

plt.title('Spettro Mercurio , $\lambda_T $ = 253.6 nm ')
plt.ylabel('Intensit√† [kCounts]')
plt.xlabel(' $\lambda_M $ [nm]')

plt.show()
```

```python
def fit_func( x , A , B ):
    return A * x + B
guess =np.array([1 , 0.7])
popt , pcov = curve_fit(fit_func , np_m , np_True , sigma=s_m , absolute_sigma = True , p0 = guess)
perr = np.sqrt(np.diag(pcov))
print(popt)
print(perr)
```

```python
x_model = np.linspace( min(np_m) , max(np_m) , 100)
y_model = fit_func (x_model , *popt )

def get_chi2_fitvalue ( x_data , y_data , yerr , model_f , popt):
    return sum((model_f(x_data,*popt) - y_data)**2 / yerr**2 )

get_chi2_fitvalue( np_m , np_True , s_m , fit_func , popt )
```
